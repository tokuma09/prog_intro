# Basis of plot 

In the textbook, author uses pylab. However, pyplot is  a more popular tool that I think. 

Here we describe basic usage.

More examples are in https://matplotlib.org/gallery.html.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()　# to use seaborn
%matplotlib inline　# to display figure inside jupyter notebook
```

```python
x = np.linspace(0.01,10.0,1000)
y = np.cos(x)
```

```python

plt.plot(x,y,label="cos")
plt.title("Cos function") #title
plt.ylim(-1.0,1.0)
plt.hlines([-0.25, 0.25],0.01,10.0, linestyles="dashed")  # plot dashed line
plt.legend() # add label
```

```python
xmin = 0.01
xmax = 10.0
numgrids = 1000
x = np.linspace(xmin,xmax,numgrids)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Subplot
plt.subplot(2, 1, 1)
plt.plot(x, y_sin,label="sin")
plt.title(r"$\sin x$")
plt.xlim(xmin,xmax)
plt.legend(loc="lower right") # select legend place

plt.subplot(2, 1, 2)
plt.plot(x, y_cos,label ="cos")
plt.title(r"$\cos x$")
plt.xlim(xmin, xmax)
plt.legend(loc="upper left")


plt.tight_layout()  # for readable title
```

```python
x = np.random.randn(30)
y = np.cos(x) + np.random.randn(30)
plt.plot(x, y, "o")  # scatter plot
```

```python
plt.plot(x, y, "ro") # change marker color
```
